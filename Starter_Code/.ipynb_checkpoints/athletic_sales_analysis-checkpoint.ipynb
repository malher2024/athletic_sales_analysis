```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_sales_2020 = pd.read_csv('./Resources/athletic_sales_2020.csv')
df_sales_2021 = pd.read_csv('./Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
display(df_sales_2020)
```

```python
# Display the 2021 sales DataFrame
display(df_sales_2021)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print("\nData types for 2020 sales data:")
# print(df_sales_2020.dtypes)
df_sales_2020.info()
```

```python
# Check the 2021 sales data types.
print("\nData types for 2021 sales data:")
# print(df_sales_2020.dtypes)
df_sales_2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_combined_sales = pd.concat([df_sales_2020, df_sales_2021], ignore_index=True)

# Reset the index
df_combined_sales = df_combined_sales.reset_index(drop=True)

print(f"Combined DataFrame shape: {df_combined_sales.shape}")

df_combined_sales.head()

```

```python
# Check if any values are null.

null_values = df_combined_sales.isnull().sum()

print(null_values)

# Optionally, check if there are any missing values overall
print(f"Total missing values: {df_combined_sales.isnull().sum().sum()}")
```

```python
# Check the data type of each column
df_combined_sales.dtypes
df_combined_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
date_format = '%Y-%m-%d'

df_combined_sales['invoice_date'] = pd.to_datetime(df_combined_sales['invoice_date'], format=date_format, errors='coerce')

df_combined_sales.info()

df_combined_sales.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_combined_sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold_summary = df_combined_sales.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

products_sold_summary = products_sold_summary.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
products_sold_summary.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table_sales = df_combined_sales.pivot_table(
    values='units_sold',            # Values to aggregate
    index=['region', 'state', 'city'],  # Index for rows
    aggfunc='sum',                  # Aggregation function
    fill_value=0                    # Replace NaN values with 0
).reset_index()


# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table_sales = pivot_table_sales.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
pivot_table_sales.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
sales_summary_groupby = df_combined_sales.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

sales_summary_groupby = sales_summary_groupby.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
sales_summary_groupby.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_sales_summary = df_combined_sales.pivot_table(
    values='total_sales',            # Values to aggregate
    index=['region', 'state', 'city'],  # Index for rows
    aggfunc='sum',                  # Aggregation function
    fill_value=0                    # Replace NaN values with 0
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_sales_summary = pivot_table_sales_summary.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
pivot_table_sales_summary.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = df_combined_sales[df_combined_sales['product'].str.contains('women\'s athletic footwear', case=False, na=False)]

womens_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_units_sold = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

womens_footwear_units_sold = womens_footwear_units_sold.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
womens_footwear_units_sold.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table_womens_footwear = womens_footwear_sales.pivot_table(
    values='units_sold',              # Values to aggregate
    index=['retailer', 'region', 'state', 'city'],  # Index for rows
    aggfunc='sum',                    # Aggregation function
    fill_value=0                      # Replace NaN values with 0
).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table_womens_footwear = pivot_table_womens_footwear.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
pivot_table_womens_footwear.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_footwear_sales = df_combined_sales[df_combined_sales['product'].str.contains('women\'s athletic footwear', case=False, na=False)]

pivot_table_sales_by_date = womens_footwear_sales.pivot_table(
    values='total_sales',         # Values to aggregate
    index='invoice_date',         # Index for rows
    aggfunc='sum',                # Aggregation function
    fill_value=0                  # Replace NaN values with 0
).reset_index()

pivot_table_sales_by_date.head()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_sales_by_date = pivot_table_sales_by_date.rename(columns={'total_sales': 'Total Sales'})


# Show the table.
pivot_table_sales_by_date.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_table_sales_by_date.set_index('invoice_date', inplace=True)
daily_sales_resampled = pivot_table_sales_by_date.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_resampled_sorted = daily_sales_resampled.sort_values(by='Total Sales', ascending=False)

daily_sales_resampled_sorted.head()
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales_resampled = pivot_table_sales_by_date.resample('W').sum()

weekly_sales_resampled_sorted = weekly_sales_resampled.sort_values(by='Total Sales', ascending=False)


# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_resampled_sorted.head()
```
